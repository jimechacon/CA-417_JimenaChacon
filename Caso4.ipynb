```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, confusion_matrix,  roc_curve, auc
import seaborn as sns
import statsmodels.formula.api as smf
```

```python


# Carga del dataset
data = pd.read_csv('train.csv')
```

Variable Dependiente: Credit Score 

```python
data['Credit_Score'].unique()
data['Credit_Score'].value_counts()
data['malo']=data['Credit_Score'].map({'Good':0, 'Standard':0, 'Poor':1})
data['malo'].value_counts()
```

Variables independientes: Edad

```python
data['Age'].unique()
```

```python
data['Edad'] = data['Age'].str.replace('_','')
data['Edad'] = pd.to_numeric(data['Edad'],errors='coerce')
data['Edad'].isna().sum()
data = data.loc[(data['Edad']<=90)&(data['Edad']>=18),:].copy()
```

```python
data['EdadCat'] = pd.cut(data['Edad'],bins = 5)
data
```

```python
data['EdadCat'].value_counts()
```

```python
dfTablaWoeEdad = data[['EdadCat','malo']].groupby('EdadCat').agg({'malo':['sum','count']})
dfTablaWoeEdad.columns = ['malo','total']
dfTablaWoeEdad['probMalo'] = dfTablaWoeEdad['malo'] / dfTablaWoeEdad['total']
dfTablaWoeEdad
```

```python
dfTablaWoeEdad['bueno'] = dfTablaWoeEdad['total']-dfTablaWoeEdad['malo']
dfTablaWoeEdad['distBad'] = dfTablaWoeEdad['malo'] / dfTablaWoeEdad['malo'].sum() 
dfTablaWoeEdad['distGood'] = dfTablaWoeEdad['bueno'] / dfTablaWoeEdad['bueno'].sum()
dfTablaWoeEdad['woe_edad'] = np.log(dfTablaWoeEdad['distBad']/dfTablaWoeEdad['distGood'])
dfTablaWoeEdad
```

```python
dfTablaWoeEdad.reset_index(inplace=True)
```

```python
data['EdadCat'] = data['EdadCat'].astype(str)
data = pd.merge(data,dfTablaWoeEdad[['EdadCat','woe_edad']],on='EdadCat')
```

```python
model = smf.logit('malo ~ woe_edad', data=data).fit()
print(model.summary())
```
